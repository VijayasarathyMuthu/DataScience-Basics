```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt
```

## Load the Data
The data is downloaded from the source and kept in the folder of the notebook and read directly from that folder using pandas library

```python
df = pd.read_csv('cell_samples.csv')
df.head()
```

The ID filed contains the reference id to the patient. The characteristics of the patient are defined as the features from Clump to Mit.

The Class field contains the diagnosis, as confirmed by separate medical procedures, as to whether the samples are benign (value = 2) or malignant (value = 4).

```python
df.shape
```

```python
ax = df[df['Class'] == 4][0:50].plot(kind= 'scatter', x= 'Clump',y= 'UnifSize', color= 'DarkBlue', label= 'malignant' );
df[df['Class']==2][0:50].plot(kind = 'scatter', x = 'Clump', y = 'UnifSize', color = 'Yellow', label = 'benign', ax=ax);
plt.show()
```

## Data pre-processing and selection

```python
df.dtypes
```

```python
df = df[pd.to_numeric(df['BareNuc'], errors= 'coerce').notnull()]
df['BareNuc'] = df['BareNuc'].astype('int64')
df.dtypes
```

```python
feature_df= df[['Clump','UnifSize','UnifShape', 'MargAdh', 'SingEpiSize', 'BareNuc', 'BlandChrom', 'NormNucl', 'Mit']]
x = np.asarray(feature_df)
x[0:10]
```

```python
df['Class']=df['Class'].astype('int')
y = np.asarray(df['Class'])
y[0:10]
```

### Train & Test Data

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 4)
print("Train set:", X_train.shape, y_train.shape)
print("Test set:", X_test.shape, y_test.shape)
```

# Modeling (SVM with Scikit-learn)

```python
#Mapping data into higher dimensional space is called Kernelling
from sklearn import svm
clf = svm.SVC(kernel= 'rbf')
clf.fit(X_train, y_train)
```

#### As it is trained. Now, lets use the model to predict using test data

```python
yhat = clf.predict(X_test)
yhat[0:10]
```

# Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[2,4])
np.set_printoptions(precision=2)

print (classification_report(y_test, yhat))
```
